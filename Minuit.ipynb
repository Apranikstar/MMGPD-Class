```python
x_data1 = xx["Mainz"]["Q2"]  # First column: x
y_data1 = xx["Mainz"]["GE"]  # Second column: y
dy_data1 = xx["Mainz"]["dGE"]  # Third column: dy
```

```python
EMOBS = MMGPD.EMObservables("NNPDF40_nlo_as_01180", mu2 = 4, mp = 0.93827208816, mn = 0.9395654133, mup = 2.7928473446 , mun = -1.9130427)
```

```python
def GEMainz(t, 
           H_aprime_uv, H_aprime_dv, H_aprime_sv,
           H_B_uv, H_B_dv, H_B_sv,
           H_A_uv, H_A_dv, H_A_sv,
           E_aprime_uv, E_aprime_dv, E_aprime_sv,
           E_B_uv, E_B_dv, E_B_sv,
           E_A_uv, E_A_dv, E_A_sv,
           E_alpha_uv, E_alpha_dv, E_alpha_sv,
           E_beta_uv, E_beta_dv, E_beta_sv,
           E_gamma_uv, E_gamma_dv,E_gamma_sv,
           Ks  
           ):
    return EMOBS.GE(1,t,
           H_aprime_uv, H_aprime_dv, H_aprime_sv,
           H_B_uv, H_B_dv, H_B_sv,
           H_A_uv, H_A_dv, H_A_sv,
           E_aprime_uv, E_aprime_dv, E_aprime_sv,
           E_B_uv, E_B_dv, E_B_sv,
           E_A_uv, E_A_dv, E_A_sv,
           E_alpha_uv, E_alpha_dv, E_alpha_sv,
           E_beta_uv, E_beta_dv, E_beta_sv,
           E_gamma_uv, E_gamma_dv,E_gamma_sv,
           Ks  
           )
```

```python
# Define the least squares for each dataset
lsq1 = LeastSquares(x_data1, y_data1, dy_data1, GEMainz)
#lsq2 = LeastSquares(x_data2, y_data2, dy_data2, GE/GD)


# Combine the least squares
least_squares = lsq1 # + lsq2



#m = Minuit(least_squares, au=6, bu=0.8, cu=4.0, ad=6, bd=0.8, cd=4.0, alpha=1)  # starting values
m = Minuit(least_squares, H_aprime_uv=0.65929631799465127,
H_aprime_dv=0.40868999246462828,
H_aprime_sv=0.0000000000000000,
H_B_uv=0,
H_B_dv=0,
H_B_sv=0.0000000000000000,
H_A_uv=0,
H_A_dv=0,
H_A_sv=0.0000000000000000,
E_aprime_uv=1.0254625137579993,
E_aprime_dv=0.74311421431656599,
E_aprime_sv=0.0000000000000000,
E_B_uv=0,
E_B_dv=0,
E_B_sv=0.0000000000000000,
E_A_uv=0,
E_A_dv=0,
E_A_sv=0.0000000000000000,
E_alpha_uv=0.48192198066489028,
E_alpha_dv=0.61964948615757021,
E_alpha_sv=0.0000000000000000,
E_beta_uv=7.1264980899423556,
E_beta_dv=17.888673527013204,
E_beta_sv=0.0000000000000000,
E_gamma_uv=0.0000000000000000,
E_gamma_dv=0.0000000000000000,
E_gamma_sv=0.0000000000000000,
Ks=0.0000000000000000)
m
```

```python
m.fixed["H_aprime_sv", "H_B_uv","H_B_dv","H_B_sv","H_A_uv" , "H_A_dv" , "H_A_sv" , "E_aprime_sv","E_B_uv" ,"E_B_dv", "E_B_sv",
"E_A_uv", "E_A_dv", "E_A_sv", "E_alpha_sv", "E_beta_sv",  "E_gamma_uv", "E_gamma_dv", "E_gamma_sv", "Ks" , "E_alpha_uv",
"E_alpha_dv","E_beta_uv" , "E_beta_dv"
] = True
```

```python

m.migrad()  # finds minimum of least_squares function
m.hesse()  # accurately computes uncertainties
```
