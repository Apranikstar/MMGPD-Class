```python
import MMGPD
import matplotlib.pyplot as plt
```

```python
x = MMGPD.GPDAnalysis("HGAG23")
```

```python
myobs = MMGPD.Observables(x)
```

```python
D0 = -0.35914486263742873
myobs.r2mass_p_w_sea(D0)
```

```python
data = -MMGPD.SkewedDataGenerator(0, 6, 200,1)()
```

```python
results_w_sea = [myobs.r2mass_p_w_sea(D0) for D0 in data]
```

```python
plt.figure(figsize=(8, 6), dpi=300)  # Corrected: Separate figsize and dpi
plt.plot(data, results_w_sea, label="r2mass vs. data", color="blue", linewidth=2)
plt.title("Plot of r2mass as a Function of D0 with Sea Quark Contribution", fontsize=12)
plt.xlabel(" D0", fontsize=12)
plt.ylabel("r2mass(D0)", fontsize=12)
plt.axhline(0, color="gray", linestyle="--", linewidth=0.8)  # Optional: y=0 line
plt.grid(True, linestyle="--", alpha=0.7)
plt.legend(fontsize=12)
plt.show()
```

```python
results_wo_sea = [myobs.r2mass_p_wo_sea(D0) for D0 in data]
```

```python
plt.figure(figsize=(8, 6), dpi=300)  # Corrected: Separate figsize and dpi
plt.plot(data, results_wo_sea, label="r2mass vs. data", color="blue", linewidth=2)
plt.title("Plot of r2mass as a Function of D0 without Sea Quark Contribution", fontsize=12)
plt.xlabel(" D0", fontsize=12)
plt.ylabel("r2mass(D0)", fontsize=12)
plt.axhline(0, color="gray", linestyle="--", linewidth=0.8)  # Optional: y=0 line
plt.grid(True, linestyle="--", alpha=0.7)
plt.legend(fontsize=12)
plt.show()
```
